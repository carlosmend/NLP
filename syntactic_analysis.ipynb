```python
!python -m spacy download es_core_news_lg
import spacy
nlp = spacy.load("es_core_news_lg")
```

```python
data_raw=pd.read_csv('data_raw.csv')
data_raw.sample(10)
```

```python
import string
doc = nlp(data_raw.traducido[4])
print(doc.text)
vec_aux=[]
for token in doc:
    if token.text not in string.punctuation:
        if token.dep_=='ROOT':
          print(token.text, " ->", token.pos_, "(", spacy.explain(token.pos_), ")","DEP--> ",token.dep_,"(", spacy.explain(token.dep_), ")")
```

```python
def sintactico(text):
  text = re.sub("[%s]" % re.escape(string.punctuation), "", text)
  if text=='':
    return "{:.3f}".format(float(0)),"{:.3f}".format(float(0)),"{:.3f}".format(float(0))
  else:
    l=len(text)
    doc=nlp(text)
    NOUN=[]
    AUX=[]
    VERB=[]
    ADJ=[]
    ADV=[]
    COUNT=[]

    for token in doc:
        if token.pos_=='NOUN':
          NOUN.append(token)

        if token.pos_=='AUX':
          AUX.append(token)

        elif token.pos_=='VERB':
          VERB.append(token)

        elif token.pos_=='ADJ':
          ADJ.append(token)

        elif token.pos_=='ADV':
          ADV.append(token)
    COUNT.append("{:.3f}".format(float(len(NOUN)/l)))
    COUNT.append("{:.3f}".format(float(len(AUX)/l)))
    COUNT.append("{:.3f}".format(float(len(VERB)/l)))
    COUNT.append("{:.3f}".format(float(len(ADJ)/l)))
    COUNT.append("{:.3f}".format(float(len(ADV)/l)))
    return COUNT[0],COUNT[1],COUNT[2],COUNT[3],COUNT[4]
```

```python
data_raw[['noun','aux', 'verb', 'adj','adv']] = data_raw['traducido'].apply(lambda x: pd.Series(sintactico(x)))

data_raw.sample(10)
```

```python
data_raw['noun']=data_raw['noun'].astype(float)
data_raw['aux']=data_raw['aux'].astype(float)
data_raw['verb']=data_raw['verb'].astype(float)
data_raw['adj']=data_raw['adj'].astype(float)
data_raw['adv']=data_raw['adv'].astype(float)
```
